`BLEU (bilingual evaluation understudy)`

**BLEU-score** — это алгоритм для оценки качества машинного перевода текста с одного естественного языка на другой.

Качество перевода оценивается путём сравнения машинного перевода с перевоом, выполненным профессиональным переводчиком. Чем ближе машинный перевод к профессиональному человеческому переводу, тем он лучше.

Показатель BLEU может иметь значение от 0 до 1.


* Что значит кандидат и референс?

**Кандидат** — это текст, полученный путём машинного перевода, т.е. тот перевод, качество которого мы хотим оценить.

**Референс** — это перевод того же текста, но выполненный профессиональным переводчиком (человеком).


Этот показатель указывает, насколько схожи текст-кандидат и текст-референс. Единица обозначает максимальное сходство канидата с референсом.


* Для чего подойдет BLEU-Score: подсчета отдельных переводов или результатов работы системы на целом корпусе?


Оценки рассчитываются для отдельных переведенных сегментов — **как правило, предложений** — путем сравнения их с набором эталонных переводов хорошего качества. Затем эти оценки усредняются по всему корпусу, чтобы получить оценку общего качества перевода.


* Какие "скоры" считаются высокими, а какие - низкими?

Высокие оценки обычно считаются значениями близкими к 1, что указывает на высокое качество перевода. Низкими оценками считаются обычно показатели, стремящиеся к нулю, они указывают на низкое качество перевода. Однако конкретные пороговые значения для определения высокого и низкого качества могут различаться в зависимости от конкретной системы оценки и задачи оценивания.

```python
import evaluate

predictions = [
    "In addition to the usual hockey, there is an underwater and even ice version of this game.", #  Помимо обычного хоккея, существует подводная и даже подлёдная разновидности этой игры.
    "You reap what you sow.", # Что посеешь - то и пожнешь.
    "Also, justice is how we distribute the small resources." # Также, справедливость - это и то, как мы распределяем малые ресурсы.
    ]

references = [
    ["In addition to conventional hockey, there is an underwater and even under-ice variety of this game."], #  Помимо обычного хоккея, существует подводная и даже подлёдная разновидности этой игры.
    ["What goes around comes around."], # Что посеешь - то и пожнешь.
    ["Fairness also has to do with how we distribute scarce resources."], # Также, справедливость - это и то, как мы распределяем малые ресурсы.
    ]

bleu = evaluate.load("bleu")
results = bleu.compute(predictions=predictions, references=references)

print(results)
```
